```python
# Q.10 Display the percentage of missing values in each column in the table labs. Show the distribution as a bar graph

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl as xl
import pandas as pd
from pywaffle import Waffle
import missingno as msno
```

```python
    Labs = pd.read_excel("G:\\My Drive\\JayanthiG\\Numpy_Ninja\\\Python\Python_Hackathon_Apr24\Cardiac_Outcomes.xlsx", sheet_name="Labs")
```

```python
Labs_missing=Labs.isna()
Labs_missing_percentage=Labs_missing.mean().round(4)*100
Labs_missing_percentage = Labs_missing_percentage[Labs_missing_percentage > 0]
Labs_missing_percentage = Labs_missing_percentage .drop(Labs_missing_percentage [Labs_missing_percentage == 0].index).sort_values(ascending=True)
Labs_missing_percentage 
```

```python
plt.figure(figsize=(6, 30))
Labs_missing_percentage.plot(kind='barh', color='green')
plt.title('Percentage of Missing Values in each Column of labs Table')
plot_width, plot_height = (16,18)
plt.rcParams['figure.figsize'] = (plot_width,plot_height)
plt.show()
```
