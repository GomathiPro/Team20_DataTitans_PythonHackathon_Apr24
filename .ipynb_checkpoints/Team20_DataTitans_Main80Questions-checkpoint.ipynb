# Q8 : Create a Waffle chart to show what percentage of total deaths belong to each admission_ward?import pandas as pd

```python
ward_grouping = {'Cardiology': 'Cardiology', 'ICU': 'ICU', 'GeneralWard': 'GeneralWard', 'Others': 'Others'}
discharge['grouped_category'] = discharge['admission_ward'].map(ward_grouping).fillna(discharge['admission_ward'])

colors = {'Cardiology': '#2196f3', 'ICU': '#ff9800', 'GeneralWard': '#4caf50', 'Others': '#87CEFA'}

discharge['percentage_death_within_28_days'] = (discharge['death_within_28_days'] / discharge['death_within_28_days'].sum()) * 100
discharge['percentage_deaths_3_months'] = (discharge['death_within_3_months'] / discharge['death_within_3_months'].sum()) * 100
discharge['percentage_death_within_6_months'] = (discharge['death_within_6_months'] / discharge['death_within_6_months'].sum()) * 100
percentage_deaths = discharge['percentage_death_within_28_days'] + discharge['percentage_deaths_3_months'] + discharge['percentage_death_within_6_months']

fig = plt.figure(
    FigureClass=Waffle, figsize=(5, 8),
    rows=6,
    columns=13,
    values=percentage_deaths,
    title={"label": "Total death percentage within categories", "loc": "center", "size": 15},
    labels=[f"{k} ({v:.1f}%)" for k, v in zip(discharge['grouped_category'], percentage_deaths)],
    colors=[colors[category] for category in discharge['grouped_category'] if category in colors],
)

legend_colors = [colors[category] for category in discharge['grouped_category'].unique()]
legend_labels = [f"{k} ({v/len(discharge)*100:.1f}%)" for k, v in discharge['grouped_category'].value_counts().items()]
legend_patches = [mpatches.Patch(color=c, label=l) for c, l in zip(legend_colors, legend_labels)]
plt.legend(handles=legend_patches, title='Legend', loc='lower center', bbox_to_anchor=(0.5, -0.6))

plt.show()
```

# # Q.24 What columns indicate a prescence of anemia? Create a severity score for anemia and prove or disprove a correlation to Length of stay

We have three column related to hemoglobin, I am taking hemoglobin column for my fin
dings.

Grade 0: Normal Hb level Grade 1: Mild anemia, 100 g/l, normal level Grade 2: Moderate anemia, 80-100 g/l Grade 3: Severe anemia, 65-80 g/l Grade 4: Life-threatening anemia, <65 g/l

Here, I am taking Grade 3: Severe anemia, 65-80 g/l to find out severity score for anemia.

Findings:There is a positive correlation between severity score for anemia and Length of stay.

```python

merged_df =pd.merge(labs,discharge,on = 'inpatient_number')

merged_df['Severity_Score'] = np.where((merged_df['hemoglobin'] >= 65) & (merged_df['hemoglobin'] <= 80), 3, 1)

Severe_Anemia = merged_df[merged_df['Severity_Score'] == 3]

print(Severe_Anemia)
sns.scatterplot(data=merged_df, x='Severity_Score', y='dischargeDay')
plt.title('Correlation between Severity Score for Anemia and Length of Stay')
plt.xlabel('Severity Score for Anemia')
plt.ylabel('Length of Stay')
plt.show()

correlation = merged_df['Severity_Score'].corr(merged_df['dischargeDay'])
print("Correlation Coefficient:", correlation)
```
